```python
import yfinance as yf

# Download GameStop stock data
gme_data = yf.download('GME', start='2010-01-01', end='2020-12-31')

# Reset index to move Date from index to column
gme_data.reset_index(inplace=True)

# Save to CSV file (optional)
gme_data.to_csv('gme_data.csv', index=False)

# Display the first five rows
print(gme_data.head())
```

```python
import requests
import pandas as pd

url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"

# Add headers to mimic a browser visit
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/90.0.4430.93 Safari/537.36"
}

response = requests.get(url, headers=headers)

# Check if request was successful
if response.status_code == 200:
    # Parse HTML tables with pandas
    gme_revenue = pd.read_html(response.text)[0]
    
    # Clean dataframe if needed
    gme_revenue.columns = ['Date', 'Revenue']
    gme_revenue = gme_revenue.dropna()
    
    # Display the last 5 rows
    print(gme_revenue.tail())
else:
    print(f"Failed to retrieve page, status code: {response.status_code}")
```

```python
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(12,6))
    plt.plot(data['Date'], data['Close'], label='Closing Price')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Assuming tesla_data is your DataFrame with Tesla stock data
make_graph(tesla_data, "Tesla Stock Closing Prices (2010-2020)")
```
